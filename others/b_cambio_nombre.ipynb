```python
import os
import pandas as pd

# Ruta de la carpeta con la extension
EXTENSION = '.jpg'
FOLDER_PATH = '/home/nahumfg/GithubProjects/parliament-voting-records/data/procesamiento_todas_votaciones/c_enderezados'

# Crear lista de archivos
png_files = [
    {'full_path': os.path.join(FOLDER_PATH, f), 'image_name': f}
    for f in os.listdir(FOLDER_PATH)
    if f.lower().endswith(EXTENSION)
]

# Crear el DataFrame
df_dir_files = pd.DataFrame(png_files)

# Crear columna 'pdf_name' dividiendo en "_page" (sin incluir) y cambiando extensión a .pdf
df_dir_files['pdf_name'] = df_dir_files['image_name'].apply(lambda x: x.split('_page')[0] + '.pdf')
df_dir_files
```

```python
import uuid

def generate_uuid_filename(url):
    full_uuid = uuid.uuid5(uuid.NAMESPACE_URL, url)
    return f"{str(full_uuid)}"

df_originales = pd.read_csv('./nombres_archivos_originales.txt', sep=";", encoding="utf-8")
df_originales = df_originales[['clean_link', 'file_name']].rename(columns={'file_name': 'pdf_name'})
df_originales['uuid_from_link'] = df_originales['clean_link'].apply(generate_uuid_filename)
df_originales
```

```python
def crear_new_image_name(row):
    import re
    match = re.search(r'_page_(\d+)\.png$', row['old_image_name'])
    if match:
        page_num = match.group(1)
        # Formato deseado: uuid_page15_.png
        return f"{row['uuid_from_link']}_page{page_num}_.png"
    else:
        return f"{row['uuid_from_link']}." + EXTENSION

df_merged = df_dir_files.merge(df_originales, on='pdf_name', how='left')
df_merged = df_merged[['full_path', 'image_name', 'uuid_from_link']]
df_merged.rename(columns={'image_name': 'old_image_name'}, inplace=True)
df_merged['new_image_name'] = df_merged.apply(crear_new_image_name, axis=1)

# Crear la columna new_full_path reemplazando old_image_name por new_image_name en el string full_path
df_merged['new_full_path'] = df_merged.apply(lambda row: row['full_path'].replace(row['old_image_name'], row['new_image_name']), axis=1)
df_merged = df_merged[['full_path', 'new_full_path','old_image_name','new_image_name']]
df_merged
```

```python
import os

# Renombra archivos de full_path a new_full_path
for idx, row in df_merged.iterrows():
    try:
        if row['full_path'] != row['new_full_path'] and os.path.exists(row['full_path']):
            os.rename(row['full_path'], row['new_full_path'])
    except Exception as e:
        print(f"Error renombrando {row['full_path']} a {row['new_full_path']}: {e}")
```
